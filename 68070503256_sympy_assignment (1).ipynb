```python
#write your code
# 1. Define symbolic variable and expression
#Define the symbolic variable 'x'
from sympy import *
x = symbols ('x')

# Define the expression E(x) = 3x^2 + 5x - 7
E_x = 3*x**2 + 5*x - 7
E_x


```

```python
# 2. Evaluate (Substitute)
# Use the .subs() method to find the numerical value of E(x) when x = 4
x_value = 4
E_value_at_4 = E_x.subs(x, x_value)
E_value_at_4
```

Part 2: Solving Equations

Tasks:

Solve a Quadratic Equation: Use the solve() function to find the exact $x$ values that satisfy:

$$\\ x^2 - 4x + 3 = 0$$

$$$$

Solve a Linear System: Define two variables, $x$ and $y$. Use solve() to find the exact values of $x$ and $y$ that satisfy both equations:

$$\\ \text{Equation 1: } x + y = 10$$

$$$$$$
$$\text{Equation 2: } 2x - y = 5

$$\\ *(Hint: Pass a list of equations and a list of variables to `solve()`.)*$$

```python
#write your code
# --- Part 2: Solving Equations ---
# 1. Solve a Quadratic Equation: x^2 - 4x + 3 = 0
# Define the variables
x, y = symbols('x y')

# Define the equation (expression set equal to zero)
quadratic_eq = x**2 - 4*x + 3

# Use solve() to find the exact x values
x_values = solve(quadratic_eq, x)

print(f"1. Quadratic Equation Solutions: x = {x_values}")
# Expected output: [1, 3]
```

```python
# 2. Solve a Linear System
# Equation 1: x + y = 10
# Equation 2: 2x - y = 5 

eq1 = x + y - 10
eq2 = 2*x - y - 5

# Use solve() to find the exact values of x and y for the system
solution = solve((eq1, eq2), (x, y))

print(f"2. Linear System Solution: x = {solution[x]}, y = {solution[y]}")
# Expected output: x = 5, y = 5
```

Tasks:

Find the Derivative: Define a function $g(x) = 4x^3 - 6x^2$. Use diff() to find the first derivative, $g'(x)$.

Find the Integral: Use integrate() to find the indefinite integral of the original function $g(x) = 4x^3 - 6x^2$. (Note: SymPy does not automatically add the constant of integration $+C$.)

```python
#write your code
# --- Part 3: Fundamental Calculus ---
# 1. Solve function g(x)
x = symbols('x')
g_x = 4*x**3 - 6*x**2
g_x
```

```python
# 1. Find the Derivative
# Use diff() to find the first derivative, g'(x)
g_prime_x = diff(g_x, x)
g_prime_x
```

```python
# 2. Find the Indefinite Integral
# Use integrate() to find the indefinite integral of g(x)
G_x = integrate(g_x, x)
G_x
```
